# 1. Source and monochromator options
The cold source operates at $T = 20K$, giving a spectrum quite different from the thermal spectrum as visualized below.
There are two different regions of the thermal spectrum that are of interest
- Around $\lambda_1 = 4$Å which would make it possible to use a Pyroletic Graphite monochromator with $\frac{d\lambda}{\lambda}=0.01$. The wavelength is roughly double of that used at a thermal source, giving a $\lambda^2$ and correspondingly $z,\tau$ that are four times as great with the same parameters otherwise. 
- Around $\lambda_2 = 6$Å, which lies outside the PG monochromator region and would require use of a velocity selector with perhaps $\frac{d\lambda}{\lambda}=0.1$. $\lambda^2$ and correspondingly $z,\tau$ would however be 9 times as great! 


Another feature of the spectrum is that the density and correspondingly the flux at $\lambda_1$ is greater than $\lambda_2$. In practice this is compensated for by the wider permitted spectrum, which in fact makes the expected flux at operating point $\lambda_2$ roughly 11 times greater! The greater $\frac{d\lambda}{\lambda}$ will however make a difference in terms of signal power at the detector, especially when scanning for higher $\delta_y$ values.

# 2. Precession devices
The three different precession devices considered are
- Isosceles triangles
- Wollaston prisms
- Magnetized foil flippers

In each case, a pair of such devices is considered as this is the minimal number needed to achieve focussing.
## 2.1. Isosceles triangles
A pair of triangles is the simplest device. Due to its simple design (aluminum coils) it can be set to quite high fields. However, $\theta_0$ is somewhat limited. They are also plagued by a wavelength-dependent phase-shift of the modulation pattern when dimensions along the neutron path do not match exactly, giving weaker modulation in practice. 

## 2.2. Wollaston prisms
Wollaston prisms are the most free of wavelength-dependent behavior. Their fields cannot be set very high though and their angle is also somewhat limited compared to foil flippers

## 2.3. Magnetized foil flippers
Magnetized foil flippers can have a very small angle and relatively high fields. However, the spin flipping of the foil is highly wavelength dependent.

## 2.4 Operating region and precession pairings
In conclusion, it appears that Wollaston prisms are the best fit for instrument designs using a velocity selector, operating at the coldest wavelength $\lambda_2$. Although the velocity detector gives a relatively broad spectrum, the lack of wavelength dependent behavior makes this just result in a narrower modulation envelope on the detector, reducing signal power. The lower max field strengths are compensated by the greater gain in $\tau$ and $\delta$ due to the high wavelength. Triangles could be too tricky to use here as they are too sensitive to imperfections in precession device dimensions. Magnetized foil flippers could also be used, although the wavelength dependency will play a bigger role. On the other hand, 

Around $\lambda_1$, triangles and foil flippers seem to make more sense.
